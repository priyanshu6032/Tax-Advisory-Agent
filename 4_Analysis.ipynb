```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the clustered dataset saved from Notebook 3
df_clustered = pd.read_csv('clustered_tax_data.csv')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))

# Generate a scatter plot mapping Gross Income against 80C Investments
sns.scatterplot(
    data=df_clustered,
    x='Gross_Income',
    y='Current_80C_Investments',
    hue='Tax_Profile_Cluster',
    palette='viridis',
    alpha=0.7 
)


plt.title('Taxpayer Profiles: Gross Income vs. Section 80C Investments')
plt.xlabel('Gross Income')
plt.ylabel('Current 80C Investments')
plt.legend(title='Profile Cluster')
plt.savefig('cluster_scatter_plot.png', bbox_inches='tight')
plt.show()


plt.figure(figsize=(10, 6))

# Generate a boxplot to show the distribution of Medical Insurance claims across the different clusters
# This highlights the variance and median behavior within each taxpayer profile
sns.boxplot(
    data=df_clustered,
    x='Tax_Profile_Cluster',
    y='Medical_Insurance',
    palette='Set2'
)

plt.title('Distribution of Medical Insurance (80D) Deductions by Profile')
plt.xlabel('Tax Profile Cluster')
plt.ylabel('Medical Insurance Premium')
plt.savefig('insurance_boxplot.png', bbox_inches='tight')
plt.show()
```
