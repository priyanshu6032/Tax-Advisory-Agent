```python
class TaxAdvisoryAgent:
    """
    An Intelligent Al agent to assist users with tax liability estimation and deductions.
    Uses rule-based modeling to simulate standard tax slabs.
    """
    def __init__(self, income, inv_80c, med_ins, charity):
        # Initialize the user's financial profile
        self.gross_income = income
        self.inv_80c = inv_80c
        self.med_ins = med_ins
        self.charity = charity

        # Define maximum allowable deductions based on standard financial rules
        self.max_80c = 150000
        self.max_med = 25000 # Assuming under 60 years old for simplicity

    def calculate_taxable_income(self):
        """Applies deduction logic to find the final taxable amount."""
        # Cap the deductions at their legal maximums
        allowed_80c = min(self.inv_80c, self.max_80c)
        allowed_med = min(self.med_ins, self.max_med)

        # Calculate total deductions
        total_deductions = allowed_80c + allowed_med + self.charity

        # Subtract deductions from gross income
        taxable_income = max(0, self.gross_income - total_deductions)
        return taxable_income, total_deductions

    def estimate_tax(self, taxable_income):
        """Rule-based tax estimation using progressive synthetic tax slabs."""
        tax = 0
        income = taxable_income

        # Slab 1: 0 to 300,000 (0% tax)
        if income > 300000:
            income -= 300000
        else:
            return tax

        # Slab 2: 300,001 to 600,000 (5% tax)
        if income > 300000:
            tax += 300000 * 0.05
            income -= 300000
        else:
            tax += income * 0.05
            return tax

        # Slab 3: 600,001 to 900,000 (10% tax)
        if income > 300000:
            tax += 300000 * 0.10
            income -= 300000
        else:
            tax += income * 0.10
            return tax

        # Slab 4: Above 900,000 (20% tax)
        tax += income * 0.20
        return tax

    def generate_recommendations(self):
        """Recommendation logic for tax optimization."""
        recommendations = []

        # Check if user has maximized their standard investments
        if self.inv_80c < self.max_80c:
            shortfall = self.max_80c - self.inv_80c
            recommendations.append(f"Invest an additional {shortfall} in 80C instruments (ELSS, PPF) to maximize deductions.")

        # Check if user has maximized medical insurance deductions
        if self.med_ins < self.max_med:
            shortfall = self.max_med - self.med_ins
            recommendations.append(f"Consider purchasing additional health insurance up to {shortfall} to save on taxes under 80D.")

        if not recommendations:
            recommendations.append("Excellent work! You have optimized your standard deductions.")

        return recommendations

    def run_agent_workflow(self):
        """Executes the full agent workflow: scenario analysis, estimation, and reporting."""
        taxable_income, total_deductions = self.calculate_taxable_income()
        estimated_tax = self.estimate_tax(taxable_income)
        advice = self.generate_recommendations()

        # Compile the final report
        report = {
            "Gross Income": self.gross_income,
            "Total Deductions Used": total_deductions,
            "Final Taxable Income": taxable_income,
            "Estimated Tax Liability": estimated_tax,
            "Agent Recommendations": advice
        }
        return report

# Instantiate the agent with a sample user's financial data
sample_user = TaxAdvisoryAgent(income=1200000, inv_80c=100000, med_ins=10000, charity=5000)

# Run the workflow and print the results
results = sample_user.run_agent_workflow()
print("--- Tax Advisory Agent Report ---")
for key, value in results.items():
    print(f"{key}: {value}")
```
