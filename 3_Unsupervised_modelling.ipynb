```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the synthetic tax data generated in Notebook 1
df_tax_data = pd.read_csv('synthetic_tax_data.csv')

# Select the relevant financial features for clustering
# We exclude 'User_ID' and 'Age' to focus strictly on financial behavior
features = ['Gross_Income', 'Current_80C_Investments', 'Medical_Insurance', 'Charity_Donations']
X = df_tax_data[features]

# Initialize the StandardScaler to standardize features by removing the mean and scaling to unit variance
# This step is critical because income is much larger than medical insurance, which would skew the K-Means distance metric
scaler = StandardScaler()

# Fit the scaler to the data and transform it into a normalized array
X_scaled = scaler.fit_transform(X)

# Initialize the KMeans clustering model
# We choose 3 clusters to represent standard profiles: e.g., Low, Medium, and High financial engagement
# random_state=42 ensures the clustering is reproducible across multiple runs
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# Fit the K-Means model to the scaled data and predict the cluster label for each user
# The resulting labels (0, 1, 2) are assigned back to the original dataframe
df_tax_data['Tax_Profile_Cluster'] = kmeans.fit_predict(X_scaled)

# Display the average financial metrics for each newly discovered cluster
# This helps us interpret what each cluster actually represents in the real world
cluster_summary = df_tax_data.groupby('Tax_Profile_Cluster')[features].mean()
print("Cluster Centers (Average values per profile):")
print(cluster_summary)

df_tax_data.to_csv('clustered_tax_data.csv', index=False)
```
