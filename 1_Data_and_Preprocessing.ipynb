```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

num_samples = 5000

# Generate synthetic data for the tax advisory agent
# Income ranges from 300,000 to 3,000,000 to cover various tax slabs
data = {
    'User_ID': range(1, num_samples + 1),
    'Age': np.random.randint(22, 65, num_samples),
    'Gross_Income': np.random.randint(300000, 3000000, num_samples),
    # Existing investments under standard sections (e.g., Section 80C in India)
    'Current_80C_Investments': np.random.randint(0, 200000, num_samples),
    # Medical insurance premiums (Section 80D)
    'Medical_Insurance': np.random.randint(0, 75000, num_samples),
    # Donations to charity (Section 80G)
    'Charity_Donations': np.random.randint(0, 50000, num_samples)
}

df_tax_data = pd.DataFrame(data)

print(df_tax_data.head())

df_tax_data.to_csv('synthetic_tax_data.csv', index=False)
```
